```python
df_northamerica = df[df.continent == 'North America']
df_northamerica
```

```python
df_europe = df[df.continent == "Europe"]
df_europe
```

```python
max_europe = df_europe.groupby('location')['total_cases'].max().sum() 
max_northamerica = df_northamerica.groupby('location')['total_cases'].max().sum()
print(max_europe, max_northamerica) #VALORE MASSIMO
```

```python
mean_europe = df_europe['total_cases'].mean().round(1)
mean_northamerica = df_northamerica['total_cases'].mean().round(1)
print(mean_europe, mean_northamerica) #MEDIA
```

```python
pulizia_df = df.dropna(subset=['location','total_cases'])
total_world= pulizia_df['new_cases'].sum()
percent_europe=(df_europe["new_cases"].sum()/total_world)*100
percent_europe.round(2) #PERCENTUALE
```

```python
pulizia_df = df.dropna(subset=['location','total_cases'])
total_world= pulizia_df['new_cases'].sum()
percent_northamerica=(df_northamerica["new_cases"].sum()/total_world)*100
percent_northamerica.round(2) #PERCENTUALE
```

```python
# Creazione DF informativo Europa

info_europe = pd.DataFrame({
    'Continent': 'Europe',
    'Max Total': [max_europe],
    'Mean': [mean_europe],
    'Percentage': [percent_europe]

}).round(2)

# Creazione DF informativo Oceania

info_northamerica = pd.DataFrame({
    'Continent': 'North America',
    'Max Total': [max_northamerica],
    'Mean': [mean_northamerica],
    'Percentage': [percent_northamerica]

}).round(2)
```

```python
unione_df = pd.concat([info_europe, info_northamerica]).reset_index(drop=True)
unione_df
```

Selezionare i dati relativi all'Italia nel 2022, e mostrare con un grafico adeguato l'evoluzione del casi totali rispetto alla data. Mostrare poi con un grafico adeguato il numero di nuovi casi rispetto alla data (filtrare i dati se necessario). Mostrare infine un grafico che mostra l'andamento della somma cumulativa nuovi casi del 2022, e commentare se ci sono similarità con l'andamento del numero dei casi totali;

```python
df["date"]= pd.to_datetime(df["date"])
```

```python
italy=df[(df.location=="Italy")&(df["date"].dt.year==2022)]
italy
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
sns.lineplot(data = italy, x = "date", y = "total_cases")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
italy.new_cases.replace(0, np.nan, inplace=True)
italy = italy.dropna(subset=["date"])
plt.figure(figsize=(20,10))
sns.lineplot(data=italy, x="date", y="new_cases")
```

```python
italy_sum = italy['new_cases'].cumsum()
plt.figure(figsize=(20,10))
sns.lineplot(data=italy, x='date', y=italy_sum)
```

```python
plt.figure(figsize=(20, 10))

sns.lineplot(data=italy, x="date", y="total_cases", label='Casi Totali')
italy_somma = italy['new_cases'].cumsum()

sns.lineplot(data=italy, x='date', y=italy_somma, label='Somma Cumulativa Nuovi Casi')

plt.xlabel('Data')
plt.ylabel('Numero di Casi')
plt.title('Confronto dei Casi Totali e Somma Cumulativa dei Nuovi Casi in Italia')

plt.legend()

plt.show()
```

Riguardo le nazioni di Italia, Germania e Francia, mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) da maggio 2022 (incluso) ad aprile 2023 (incluso), e scrivere un breve commento a riguardo;

```python
ita_ger_fra = df[df['location'].isin(['Italy', 'Germany', 'France'])]
ita_ger_fra = ita_ger_fra[(ita_ger_fra['date'] >= '2022-05-01') & (ita_ger_fra['date'] <= '2023-04-30')]
ita_ger_fra
```
